```python
'''
Q-7> What is a heatmap, and when should it be used ?
Sol:
A heatmap is a data visualization technique that represents values using color gradients.It is commonly used to show the magnitude of data.
Ex:
'''
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {'Population-Place-Status': ['Punjab', 'Banglore', 'Pune', 'Harayana', 'Vizag'],'Population-in-Percentage': ['54%', '87%', '72%', '88%', '67%'],  # Fixing incorrect string syntax
'Growth-Rate': [12.5, 7.9, 9, 4.2, 4.0],'Death-rate': [1.1, 3.0, 6.5, 0.9, 0.2]}
df = pd.DataFrame(data)
df['Population-in-Percentage'] = df['Population-in-Percentage'].str.replace('%', '').astype(float)
mat = df.drop(columns=['Population-Place-Status']).corr()
plt.figure(figsize=(7, 4))
sns.heatmap(mat, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("****************************Census Data for 2024-2025**************************")
plt.show()

```

```python
'''
Q-20> What is the role of a pivot table in Pandas ?
Sol:
Pivot table in Pandas is used to summarize and analyze data by grouping, aggregating, and reshaping it in a structured format.
Ex:
'''
import pandas as pd
data = { "City": ["Kolkata", "Vizag", "Gurugram", "Gujrat", "New Delhi"],"Category": ["Audi", "Porche", "Thar", "Laborgini","BMW"],
"Sales": [123,435,690,700,50]}
df = pd.DataFrame(data)
print(df)
print("-------------------------------Above Table Details--------------------------------------------")
pivot_of_the_above_table = pd.pivot_table(df, values="Sales", index="City", columns="Category", aggfunc="sum")
print(pivot_of_the_above_table)
print("---------------------------------------Pivot Table to Find Total Sales by City & Category-------------------------------------------------------")
pivot_of_the_above_table1 = pd.pivot_table(df, values="Sales", index="City", columns="Category", aggfunc="sum")
print(pivot_of_the_above_table1)
print("--------------------------------------Multiple Aggregation Functions---------------------")
pivot_of_the_above_table2 = pd.pivot_table(df, values="Sales", index="City", columns="Category", aggfunc=["sum", "mean"])
print(pivot_of_the_above_table2)
print("---------------------------------------------Multiple Aggregation Functions--------------------")
pivot_of_the_above_table3 = pd.pivot_table(df, values="Sales", index="City", columns="Category", aggfunc=["sum", "mean"])
print(pivot_of_the_above_table3)

```
